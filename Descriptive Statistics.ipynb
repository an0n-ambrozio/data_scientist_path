```python
# Exercise 1: Basic Descriptive Statistics

# a) Load the "Iris" dataset into a Pandas DataFrame.
# b) Use the describe() method to get a summary of the descriptive statistics for all numerical columns.
# c) Print the mean, median, and standard deviation for the "sepal_length" column.
```

```python
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
print(iris.head())
```

```python
import pandas as pd
```

```python
print(iris.describe())
```

```python
print("Mean of sepal_length:", iris['sepal_length'].mean()) # Display the average value
print("Median of sepal_length:", iris['sepal_length'].median()) # Display the middle value when data is sorted.
print("Standard Deviation of sepal_length:", iris['sepal_length'].std()) # Display the square root of the variance
```

```python
print("Mean of petal_length:", iris['petal_length'].mean())
print("Median of petal_length:", iris['petal_length'].median())
print("Standard Deviation of petal_length:", iris['petal_length'].std())
```

```python
# Exercise 2: Measures of Variability

# Calculate the range for the "petal_width" column.
# Calculate the variance for the "sepal_width" column.
# Calculate the interquartile range (IQR) for the "petal_length" column.
```

```python
petal_width_range = iris['petal_width'].max() - iris['petal_width'].min()
print("Range of petal_width:", petal_width_range)
```

```python
sepal_width_range = iris['sepal_width'].max() - iris['sepal_width'].min()
print("Range of sepal_width:", sepal_width_range)
```

```python
sepal_length_variance = iris['sepal_length'].var()
print("Variance of sepal_length:", sepal_length_variance)
```

```python
sepal_width_variance = iris['sepal_width'].var()
print("Variance of sepal_width:", sepal_width_variance)
```

```python
petal_length_interquatile = iris['petal_length'].quantile()
print("Interquatile of petal_length:", petal_length_interquatile)
```

```python
petal_length_q75 = iris['petal_length'].quantile(0.75)
petal_length_q25 = iris['petal_length'].quantile(0.25)
petal_length_IQR = petal_length_q75 - petal_length_q25
print(petal_length_q75)
print(petal_length_q25)
print("The IQR for petal_length is:", petal_length_IQR)
```

```python
# Exercise 3: Grouped Statistics

# Calculate the mean "sepal_length" for each species of iris.
# Calculate the median "petal_width" for each species of iris.
# Calculate the standard deviation of "sepal_width" for each species of iris.
```

```python
iris.groupby('species').head()
```

```python
species_names = iris['species']
print(species_names)
```

```python
mean_sepal_length_by_species = iris.groupby('species')['sepal_length'].mean()
print("Mean sepal_length by species:", mean_sepal_length_by_species)
```

```python
mean_sepal_width_by_species = iris.groupby('species')['sepal_width'].mean()
print("Mean sepal_width by species:", mean_sepal_width_by_species)
```

```python
median_sepal_length_by_species = iris.groupby('species')['sepal_length'].median()
print("Median sepal_length by species:", median_sepal_length_by_species)
```

```python
median_sepal_width_by_species = iris.groupby('species')['sepal_width'].median()
print("Median sepal_width by species:", median_sepal_width_by_species)
```

```python
std_sepal_length_by_species = iris.groupby('species')['sepal_length'].std()
print("The Standard Deviations for each species are:", std_sepal_length_by_species)
```

```python
# Exercise 4: Measures of Shape

# Calculate the skewness for the "sepal_length" column.
# Calculate the kurtosis for the "petal_width" column.
```

```python
sepal_length_skewness = iris['sepal_length'].skew()
print("Skewness of sepal_length:", sepal_length_skewness)
```

```python
petal_width_kurtosis = iris['petal_width'].kurtosis()
print("Kurtosis of petal_width:", petal_width_kurtosis)
```
